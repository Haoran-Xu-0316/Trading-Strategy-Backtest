## 1.4 脑子：我会SMA策略了！

```python
class SMAStrategy(bt.Strategy):
    params = (
        ('maperiod', 15),
    )
    def log(self, txt, dt=None):
        dt = dt or self.datas[0].datetime.date(0)
        print('%s, %s' % (dt.isoformat(), txt))
        
    def __init__(self):
        self.dataclose = self.datas[0].close
        # 添加移动平均线指标        
        self.sma = bt.indicators.SimpleMovingAverage(self.datas[0], period=self.params.maperiod) 
        #  self.sma = bt.talib.SMA(self.datas[0], timeperiod=self.params.maperiod)
        self.buy_signal = bt.indicators.CrossOver(self.dataclose, self.sma)
        self.sell_signal = bt.indicators.CrossDown(self.dataclose, self.sma)  
        
    def start(self):
         print("策略启动！")  
            
    def prenext(self):
        print("再等等...指标还没生成")
        
    def next(self):
        self.log('收盘价: %.2f' % self.dataclose[0])
        
        if self.buy_signal[0] == 1:
                #执行买入操作
                self.log('买入: %.2f' % self.dataclose[0])
                #调用Oders模块
                self.buy()
                # Create a buy (long) order and send it to the broker
        else:
            
            if self.sell_signal[0] == 1:
                # 如果持仓，当收盘价小于平均线，则卖出
                self.log('卖出, %.2f' % self.dataclose[0])
                self.sell()   
        
    def stop(self):
        print("策略回测完毕")
    
cerebro = bt.Cerebro()

#加载数据
file_name = '600600_daily_2021.csv'
data_raw = pd.DataFrame(pd.read_csv(file_name, index_col=0,parse_dates=True)).dropna()

#创建 Data Feed
data = bt.feeds.PandasData(
        dataname=data_raw,
        # 回测起始时间
        fromdate=datetime.datetime(2017, 1, 1),
        # 回测结束时间
        todate=datetime.datetime(2021, 12, 31),)

# 将Data Feed添加到Cerebro
cerebro.adddata(data)

# 添加策略
cerebro.addstrategy(SMAStrategy)

print('账户启动资金: %.2f' % cerebro.broker.getvalue())
cerebro.run()
print('账户最终金额: %.2f' % cerebro.broker.getvalue())
```

```python
class SMAStrategy_V1(bt.Strategy):
    params = (
        ('maperiod', 15),
    )
    def log(self, txt, dt=None):
        dt = dt or self.datas[0].datetime.date(0)
        print('%s, %s' % (dt.isoformat(), txt))
        
    def __init__(self):
        self.dataclose = self.datas[0].close
        # 添加移动平均线指标        
        self.sma = bt.indicators.SimpleMovingAverage(self.datas[0], period=self.params.maperiod) 
        self.buy_signal = bt.indicators.CrossOver(self.dataclose, self.sma)
        self.sell_signal = bt.indicators.CrossDown(self.dataclose, self.sma)   
        # 跟踪挂单
        self.order = None
    
    def notify_order(self, order):
        if order.status in [order.Submitted, order.Accepted]:
            return

        if order.status in [order.Completed]:
            if order.isbuy():
                self.log('买入成交: %.2f' % order.executed.price)
            elif order.issell():
                self.log('卖出成交: %.2f' % order.executed.price)
       
        elif order.status in [order.Canceled, order.Margin, order.Rejected]:
            self.log('订单被取消/拒绝')

        self.order = None
        
    def notify_trade(self, trade):
        if not trade.isclosed:
            return

        self.log('交易收益： 总收益 %.2f, 净收益 %.2f' %
                 (trade.pnl, trade.pnlcomm))    
        
    def start(self):
         print("策略启动！")  
        
    def next(self):
        self.log('收盘价: %.2f' % self.dataclose[0])
        if self.order:
            return
        # 检查我们是否处于空仓状态
        if not self.position:        
            if self.buy_signal[0] == 1:

                    self.log('买入: %.2f' % self.dataclose[0])
                    self.order = self.buy()

        else:            
            if self.sell_signal[0] == 1:

                self.log('卖出, %.2f' % self.dataclose[0])
                self.order = self.sell()  
                
    def stop(self):
        print("策略回测完毕")
    
cerebro = bt.Cerebro()

#加载数据
file_name = '600600_daily_2021.csv'
data_raw = pd.DataFrame(pd.read_csv(file_name, index_col=0,parse_dates=True)).dropna()

#创建 Data Feed
data = bt.feeds.PandasData(
        dataname=data_raw,
        # 回测起始时间
        fromdate=datetime.datetime(2017, 1, 1),
        # 回测结束时间
        todate=datetime.datetime(2021, 12, 31),)

# 将Data Feed添加到Cerebro
cerebro.adddata(data)

# 添加策略
cerebro.addstrategy(SMAStrategy_V1)

print('账户启动资金: %.2f' % cerebro.broker.getvalue())
cerebro.run()
print('账户最终金额: %.2f' % cerebro.broker.getvalue())
```

```python
class SMAStrategy_V2(bt.Strategy):
    params = (
        ('maperiod', 15),
    )
    def log(self, txt, dt=None):
        dt = dt or self.datas[0].datetime.date(0)
        print('%s, %s' % (dt.isoformat(), txt))
        
    def __init__(self):
        self.dataclose = self.datas[0].close
        # 添加移动平均线指标        
        self.sma = bt.indicators.SimpleMovingAverage(self.datas[0], period=self.params.maperiod) 
        self.buy_signal = bt.indicators.CrossOver(self.dataclose, self.sma)
        self.sell_signal = bt.indicators.CrossDown(self.dataclose, self.sma)  
        # 跟踪挂单
        self.order = None
    
    def notify_order(self, order):
        if order.status in [order.Submitted, order.Accepted]:
            return

        if order.status in [order.Completed]:
            if order.isbuy():
                self.log(
                    '买入成交, 成交价: %.2f, 交易成本: %.2f, 佣金: %.2f' %
                    (order.executed.price,
                     order.executed.value,
                     order.executed.comm))
          
            else:
                self.log('卖出成交, 成交价: %.2f, 交易成本: %.2f, 佣金: %.2f' %
                         (order.executed.price,
                          order.executed.value,
                          order.executed.comm))

        elif order.status in [order.Canceled, order.Margin, order.Rejected]:
            self.log('订单被取消/拒绝')

        self.order = None    
        
    def notify_trade(self, trade):
        if not trade.isclosed:
            return

        self.log('交易收益： 总收益 %.2f, 净收益 %.2f' %
                 (trade.pnl, trade.pnlcomm))    
        
    def start(self):
         print("策略启动！")  
        
    def next(self):
        self.log('收盘价: %.2f' % self.dataclose[0])
        if self.order:
            return
        # 检查我们是否处于空仓状态
        if not self.position:        
            if self.buy_signal[0] == 1:
 
                    self.log('买入: %.2f' % self.dataclose[0])
                    self.order = self.buy()

        else:            
            if self.sell_signal[0] == 1:

                self.log('卖出, %.2f' % self.dataclose[0])
                self.order = self.sell()  
                
    def stop(self):
        print("策略回测完毕")
    
cerebro = bt.Cerebro()

#加载数据
file_name = '600600_daily_2021.csv'
data_raw = pd.DataFrame(pd.read_csv(file_name, index_col=0,parse_dates=True)).dropna()

#创建 Data Feed
data = bt.feeds.PandasData(
        dataname=data_raw,
        # 回测起始时间
        fromdate=datetime.datetime(2017, 1, 1),
        # 回测结束时间
        todate=datetime.datetime(2021, 12, 31),)

# 将Data Feed添加到Cerebro
cerebro.adddata(data)

# 添加策略
cerebro.addstrategy(SMAStrategy_V2)

cerebro.addsizer(bt.sizers.FixedSize, stake=100)

cerebro.broker.setcommission(commission=0.001)
#当margin为其他值时，commission的值是一个固定价格，表示每执行一个单位合约交易所需要支付的佣金

print('账户启动资金: %.2f' % cerebro.broker.getvalue())
cerebro.run()
print('账户最终金额: %.2f' % cerebro.broker.getvalue())
```

#  2. 双均线策略构建

## 2.1 导入相关库

```python
import numpy as np
import pandas as pd
from pylab import plt, mpl

import warnings 
warnings.filterwarnings('ignore')
```

## 2.2 导入回测数据至dataframe

```python
path = 'https://gitee.com/yanzhang2021/data/raw/master/Bloomberg/EURUSD_daily.csv'
symbol = 'Close'
data = pd.DataFrame(pd.read_csv(path, index_col=0,parse_dates=True).dropna()[symbol])
data
```

```python
data=data["2000-01-01":"2005-12-31"]
data
```

## 2.3 编写相关指标及可视化显示

```python
n1=20
n2=120

# rolling(n)滑动窗口

data['SMA1'] = data[symbol].rolling(n1).mean()
data['SMA2'] = data[symbol].rolling(n2).mean()

%matplotlib widget
data.plot(figsize=(10, 8));
```

```python
data.dropna(inplace=True)

#交易信号
# 创建新列p，短期均线在长期之上时，p=1 否则p=-1
# 注意：p计算出来的依据是SMA1和SMA2，而SMA1和SMA2计算依据是当天的Close，也就是当天收盘价以后计算出的结果
data['p'] = np.where(data['SMA1'] > data['SMA2'], 1, -1)
data
```

```python
# shift(1) 将p列向下“串”一行，以求回报率;shift(-1)将某列向上串一行
#p列同时表示“明天”做多+1还是做空-1。具体的逻辑是根据“今天”收盘价计算出来的SMA1和SMA2，如果SMA1>SMA2,上穿，则明日开始做多
#这个判断应该是在“今日”的收盘价确定后，也就是收盘以后马上计算得出，而不能在是后盘以前得出。因为及时接近收盘，收盘价仍有可能变化
#只有收盘后，价格才确定不变

data['p'] = data['p'].shift(1)
data
```

```python
data.dropna(inplace=True)

#secondary_y 表示第二个坐标轴
data.plot(figsize=(10, 8), secondary_y='p');
```

## 2.4 编写相关回报计量方法

```python
# 计算“今日”的回报率，r=log(today's close/yesterday's close), data[symbol].shift(1)向下串一行，成为“昨日”的收盘价
data['r'] = np.log(data[symbol] / data[symbol].shift(1))

data
```

```python
data.dropna(inplace=True)

#p列表示1或者-1。1即将对应做多，-1对应做空。乘以回报率表示做多的收益率和做空的收益率，表示当日做单后的回报率
#默认无任何手续费以及其他成本
data['s'] = data['p'] * data['r']

data
```

```python
# r列表示当日市场给定的回报率，s列表示当日加入做单方向的回报率，取对数值
data[['r', 's']].sum().apply(np.exp)
```

```python
#取r、s两列的累积求和
data[['r', 's']].cumsum().apply(np.exp).plot(figsize=(10, 6));
```

## 2.5 其他细节考量

```python
#设置一个比例手续费，proportional cost，回报率-手续费比率=收入 
pc = 0.005
```

```python
data['s_'] = np.where(data['p'].diff() != 0,data['s'] - pc, data['s'])


#回测数据第一行处进场，s_表示考量过手续费的做单回报率
data['s_'].iloc[0] -= pc

#回测数据最后一行关单收场
data['s_'].iloc[-1] -= pc


#对dataframe进行切片，列出所有进出场位置
data[['p','r', 's', 's_']][data['p'].diff() != 0]
```

```python
#三列r,s,s_分别表示每天实际回报率，不考量手续费的做单回报率，以及考量手续费的做单回报率
# apply(np.exp)的用意是，由于r已经取对数，取对数是为了计算方便，
# 但是实际使用简单回报率更直观，所以此处使用np.exp将对数回报率转化成简单回报率
data[['r', 's', 's_']].sum().apply(np.exp)
```

```python
data[['r', 's', 's_']].cumsum().apply(np.exp).plot(figsize=(10, 6));
```
